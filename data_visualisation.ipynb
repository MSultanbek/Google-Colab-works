```python
data = pd.read_csv("/content/gpa_iq.csv")
data
```

# Data visualisation is the main part of the data analysis. Through visualisation we can understand the data more clearly, and therefore work more efficiently. There are a lot of libraries, but we will work with matplotlib, since it's very popular and still workable library for data visualisation.

## **Plot**


```python
#let's work with dataset of students grade and their iq
plt.plot(data["iq"])
plt.ylabel("iq")
plt.show()
```

"plt.plot" is the most basic graph of one or two variables in dataset. for example in the graph above we can see graph of the students iq at the y value, and students order

```python
#we can add second variable, for example let's add their grade at the y value
plt.plot(data["gpa"], data["iq"])
plt.ylabel("iq")
plt.xlabel("gpa")
plt.show()
```

we can see that it becomes a complete mess, so let's try to plot them apart


```python
plt.plot(data["gpa"])
plt.plot(data["iq"])
plt.legend(["gpa", "iq"])
plt.show()
```

actually, if we look close enough, we can see that if iq graph falls, gpa graph will also have a little groove

```python
#we can diveide them into individual subplots by plt.subplot
plt.subplot(2,1,1)
plt.plot(data["gpa"])
plt.ylabel("gpa")
plt.subplot(2,1,2)
plt.plot(data["iq"])
plt.ylabel("iq")
plt.show()
```

by using parameters we can do a lot with the plot


*   ls - stands for linestyle

    1.   "solid" or "-" is just line
    2.   "dotted" or ":"
    3. "dashed" or "--"
    4. "dashdot" or "-."

*   color - there are 140 word values for colors, we can also write our own color by hexadecimal code
* lw - stands for linewidth. we can change line width of the line in graph

by using plt.grid() we can add grid to the graph. parameters of the plt.plot also can be used here.





```python
plt.plot(data["iq"], ls = ":", color = "hotpink", lw = 5)
plt.ylabel("iq")
plt.grid()
plt.show()
```

## **Scatter**
scatter is a plot that need two arrays of the same length, for x and y value, and plot the points on the graph

```python
plt.scatter(data["iq"],data["gpa"],c = data["obs"], cmap = "viridis")
plt.ylabel("gpa")
plt.xlabel("iq")
plt.colorbar()
plt.show()
```

cmap - parameter to separate dots from one another, for example in the graph above we can see that point on (103, 0.4) is somewhere in the green part of the colorbar. There are A LOT of colormaps, you can choose whatever you want.
plt.colorbar() - function to show that thing on the right side

OTHER PARAMETER OF THE SCATTER:
* s - stands for sizes. we can give individual size for every dot in the graph.
* alpha - useful to change transparency of dots.

```python
#let's see what we can do with this parameters
colors = np.random.randint(78, size=(78))
sizes = 10 * np.random.randint(78, size=(78))
plt.scatter(data["iq"], data["gpa"], c = colors, s = sizes, alpha = 0.5, cmap = "cool")
plt.xlabel("iq")
plt.ylabel("gpa")
plt.colorbar()
plt.show()
```

## **Bar**
bar is graph with two values for x and y value

## **Histograms**

A histogram is a graph used to represent the frequency distribution of a few data points of one variable

we use plt.hist() function to create histogram

```python
#let's see example
plt.subplot(1,2,1)
plt.hist(data["iq"])
plt.title("iq")

plt.subplot(1,2,2)
x_axis = data["iq"]
mean = statistics.mean(x_axis)
sd = statistics.stdev(x_axis)
plt.title("iq")
plt.plot(x_axis, norm.pdf(x_axis, mean, sd))
plt.show()
```

normal distribution graph and histogram tells the same thing, we just need to choose what to use

as expected iq has normal distribution, which means that the values in center are more than in borders.

by this example we can see that 3 people have iq from 70 to 80, 1 student has iq from 80 to 85, 4 people - from 85 to 91 etc.

## **Pie Charts**

## **Box-plot**

Boxplot is used to summarize some data, understand properties, such as minimum, median, quartiles etc. We use plt.boxplot() function to plot boxplot

**parameters**:
* **data** - array or sequence of array to be plotted
* **notch** - optional parameter accepts boolean values
* **vert** - optional parameter accepts boolean values false and true for horizontal and vertical plot respectively
* **bootstrap** - optional parameter accepts int specifies intervals around notched boxplots
* **usermedians**	- optional parameter accepts array or sequence of array dimension compatible with data
* **positions**	- optional parameter accepts array and sets the position of boxes
* **widths** - optional parameter accepts array and sets the width of boxes
* **patch_artist** - optional parameter having boolean values
* **labels** - sequence of strings sets label for each dataset
* **meanline** - optional having boolean value try to render meanline as full width of box
* **order**	- optional parameter sets the order of the boxplot

```python
plt.boxplot(data["iq"], vert = False)
plt.title("iq")
plt.show()

x_axis = data["iq"]
mean = statistics.mean(x_axis)
sd = statistics.stdev(x_axis)
plt.title("iq")
plt.plot(x_axis, norm.pdf(x_axis, mean, sd))
plt.show()
```

let's analyze what can we get from boxplot above

4 values at the left are outliers-values which are in abnormal distance from the rest of the values.

First vertical line means minimum of the data.
In our example minimum value is somewhere between 80 to 90 (probably 87)

rectangle means values from 1st to 3rd quartile (25 to 75 percentile). In our example we can see that 1st quartile is ≈103 and 3rd quartile is ≈119.

orange line at the middle of the rectangle means median. In our example it's 110

vertical line at the right means max value. Here it's ≈138

box is a little bit righter because values are skewed a little bit to the right. From the graph lower we can see it by the normal distribution graph

```python
plt.subplot(2,1,1)
plt.boxplot(data["iq"])
plt.title("iq")

plt.subplot(2,1,2)
plt.title("gpa")
plt.boxplot(data["gpa"])
plt.show()
```

## **2D heatmap**

2d heatmap is a graph that shows the magnitude of the matrix in form of a colored table

we can use it to understand correlation between values in dataset

```python
d = np.random.random((10, 10))
plt.imshow(d, cmap="cool")
plt.title("2D heatmap")
plt.colorbar()
plt.show()
```

```python
colors_list = ['#000000', '#ffffff']
cmap = colors.ListedColormap(colors_list)
plt.imshow(d, cmap = cmap)
plt.colorbar()
plt.show()
```

we can also define our own colormap by using "cmap = colors.ListedColormap(colors_list)" method



## **Stacked bar**

Stacked bar is just bars with other bars on the top

```python
#my dataset is also not usable for showing this plots so i'll use new one
#for example
newx = [1,2,3,4]
newy1 = [10,20,30,40]
newy2 = [40,30,20,10]
dataset = pd.DataFrame([['1', 10, 20, 10, 26], ['2', 20, 25, 15, 21], ['3', 12, 15, 19, 6]],
                       columns = ["0", "1", "2", "3", "4"])

plt.bar(newx, newy1)
plt.bar(newx, newy2, bottom = newy1)
plt.show()
```

```python
#we also can add legend and also give our colors
plt.bar(newx, newy1, color = "k")
plt.bar(newx, newy2, bottom = newy1, color = "r")
plt.legend(["newy1", "newy2"])
plt.show()
```
